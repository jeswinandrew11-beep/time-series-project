```python
scaler = MinMaxScaler()
scaler.fit(train_data)
train_scaled = scaler.transform(train_data)
val_scaled = scaler.transform(val_data)
test_scaled = scaler.transform(test_data)
train_scaled_df = pd.DataFrame(train_scaled, columns=train_data.columns)
val_scaled_df = pd.DataFrame(val_scaled, columns=train_data.columns)
test_scaled_df = pd.DataFrame(test_scaled, columns=train_data.columns)
```

```python
def create_seasonal_features(df6, look_back, period=365):
    X, y = [], []
    
    for i in range(len(df6) - look_back):
        X.append(df6.iloc[i:i + look_back][['count_of_week_acc_to_year','Year','days_num','weekday_or_weekends','count_of_week_acc_to_month','Holiday(canada)','holiday(italy)','holiday(norway)','holiday(finland)','holiday(kenya)','holiday(singapore)','country_Canada','country_Finland', 'country_Italy', 'country_Kenya', 'country_Norway','country_Singapore', 'store_Discount Stickers','store_Premium Sticker Mart', 'store_Stickers for Less','product_Holographic Goose', 'product_Kaggle', 'product_Kaggle Tiers','product_Kerneler', 'product_Kerneler Dark Mode','Day_Friday','Day_Monday','Day_Saturday','Day_Sunday','Day_Thursday','Day_Tuesday','Day_Wednesday','Month_Apr','Month_Aug','Month_Dec','Month_Feb','Month_Jan','Month_Jul','Month_Jun','Month_Mar','Month_May','Month_Nov','Month_Oct','Month_Sep']].values.flatten())
        y.append(df6.iloc[i + look_back]['num_sold'])
    
    return np.array(X), np.array(y)
```

```python
look_back = 15
X_train, y_train = create_seasonal_features(train_scaled_df, look_back, period=365)
X_val, y_val = create_seasonal_features(val_scaled_df, look_back, period=365)
X_test, y_test = create_seasonal_features(test_scaled_df, look_back, period=365)
```

```python
len(X_train)
```

```python
y_train[0]
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
def build_sann_model(input_shape):
    model = Sequential()
    model.add(Dense(64, input_dim=input_shape[1], activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1)) 
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model = build_sann_model(X_train.shape)
model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=1)
```

```python
model.summary()
```

```python
y_pred1 = model.predict(X_train)
y_pred = model.predict(X_val)
y_pred2 = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,accuracy_score
from sklearn.metrics import r2_score
import math
```

```python
def calculate_rmse(y_true, y_pred):
    return math.sqrt(mean_squared_error(y_true, y_pred))
```

```python
rmsc = calculate_rmse(y_test, y_pred2)
r2 = r2_score(y_test,y_pred2)
rmsc,r2
```
